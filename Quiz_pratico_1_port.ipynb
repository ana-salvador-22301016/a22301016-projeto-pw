3 - Na questão 2, analisámos tanto o histograma como as informações descritivas das duas colunas, `year` e `mpg`. Agora, utilize o intervalo de dados observado e crie uma nova variável binária que classifique os carros como **mais antigos** ou **mais recentes**. Explique o critério escolhido.

```python

# A maioria dos carros se concentra entre os anos 76 e 80. Então defini carro com fabricação com ano de fabricação menor ou igual a 76 como "0" e maiores que 76 como "1".


df['ano_categoria'] = np.where(df['year'] <= 76, '0', '1')
# Aqui vi apenas o numero de carros .
print(df['ano_categoria'].value_counts())
print(df[['year', 'ano_categoria']])

```

```python
df.boxplot(column='year', by='ano_categoria')
plt.show()
#O plot que gerei representa dois plots que o da direita representa os carros com ano de fabrico mais antigos enquando o da esquerda apresenta os mais recentes, nota-se que há menos carros mais recentes em relação os mais antigos, podemos observar pela mediana
```

7 - Para complementar a análise acima, crie dois diagramas de dispersão separados que relacionem o consumo (mpg) e a potência (horsepower). Um dos diagramas deve conter apenas carros com 4 cilindros e o outro apenas carros com 8 cilindros.

```python
# Primeiro tive que retirar os valores '?' que eram diferente de inteiro, para isso filtrei os valores diferentes de ? e atribui ao horsepower
filtro = df['horsepower'] != '?'
df = df[filtro]
df['horsepower'] = df['horsepower'].astype(int)
```
